```python
def shellSort(array, n):

    # Rearrange elements at each n/2, n/4, n/8, ... intervals
    print('')
    print('array:',array)
    print('n:',n)
    interval = n // 2
    print('interval:',interval)
    print('------')
    while interval > 0:
        for i in range(interval, n):
            temp = array[i]
            print('     i:',i)
            print('     temp = array[i]:',temp)
            j = i
            print('     j = i:',j)
            while j >= interval and array[j - interval] > temp:
                print('             while j >= interval and array[%d] > temp'%(j-interval))
                array[j] = array[j - interval]
                print('             array[%d] = array[%d] = %d'%(j, j-interval, array[j - interval]))
                j -= interval
                print('             j - %d ='%interval, j)

            array[j] = temp
            print('             array[%d] = temp_%d'%(j,temp))
            print('             array:',array)
            print('')
        interval //= 2
        print('interval:',interval)
        print('------')


data = [9, 8, 3, 7, 5, 6, 4, 1]
size = len(data)
shellSort(data, size)
print('')
print('Sorted Array in Ascending Order:')
print(data)
```
