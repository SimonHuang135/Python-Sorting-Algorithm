```python
def bubbleSort(arr):
    n = len(arr)
    print(n)
 
    for i in range(n):
        print(' \n-----\n', 'i:', i, '\n', '')
        

        for j in range(0, n-i-1):
            print('  -------\n', '  j:', j)

            if arr[j] > arr[j+1] :
                print('     if loop:')
                print('     arr[j]', arr[j])
                print('     arr[j+1]', arr[j+1])
                arr[j], arr[j+1] = arr[j+1], arr[j]
                print('     arr[j] > arr[j+1], so swap them')
                print('         arr[j]', arr[j])
                print('         arr[j+1]', arr[j+1])
                print('')
            else:
                print('     swap done for one round')
 
arr = [64, 34, 25, 12, 22, 11, 90]
 
bubbleSort(arr)
 
print ("ordered:")
#orderedarr = []
for i in range(len(arr)):
    print ("%d" %arr[i])
    #orderedarr.append(arr[i])

#print (orderedarr)
```
