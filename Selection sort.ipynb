```python
'''
64  25  12  22  11

i=0|         A[min_index]=A[0]=64
   | j=0 |   if A[min_index]>A[1], 64>25, index_of_A[1] = min_index = 1
   | j=1 |   if A[min_index]>A[2], 25>12, index_of_A[2] = min_index = 2
   | j=2 |   if A[min_index]>A[3], 12>22
   | j=3 |   if A[min_index]>A[4], 12>11, index_of_A[4] = min_index = 4
     swap A[0] and A[min_index] => A[4]
     A = 11 25 12 22 64

i=1|         A[min_index]=A[1]=25
   | j=0 |   if A[min_index]>A[2], 25>12, index_of_A[2] = min_index = 2
   | j=1 |   if A[min_index]>A[3], 12>22
   | j=2 |   if A[min_index]>A[4], 12>64
     swap A[1] and A[min_index] => A[2]
     A = 11 12 25 22 64
     .........
'''
```

```python
A = [64, 25, 12, 22, 11] 
  
for i in range(len(A)):
    print('') 
    print('i:',i)
    print('A[i]',A[i])
    min_idx = i 
    print('A[min_idx]',A[min_idx])
    print('------------')

    for j in range(i+1, len(A)): 
        print('')
        print('j:',j)
        print('A[min_idx]= A[%d]'%min_idx,A[min_idx])
        print('A[j] = A[%d]'%j,A[j])
        print('') 

        if A[min_idx] > A[j]: 
            print('     if A[%d] > A[%d] loop'%(min_idx,j))
            min_idx = j 
            print('     min_idx = j:',min_idx)
            print('     A[min_idx = j] = A[%d]:'%min_idx,A[min_idx])
            print('     ------------')
    print('')
    print('swap position')
    A[i], A[min_idx] = A[min_idx], A[i]
    print('A[min_idx]',A[min_idx])
    print('A[i]',A[i])
    print(A)
```
